## Twitter API Challenge

Create a mongo collection of tweets about something (anything you choose).

Each mongo document should contain the text, username, favorite count and retweet count of the tweet.

```python
import requests
from requests_oauthlib import OAuth1
from pymongo import MongoClient
import tweepy
import cnfg
```

```python
client = MongoClient()
```

```python
config = cnfg.load(".twitter_config")
```

```python
auth = tweepy.OAuthHandler(config['consumer_key'],
              config['consumer_secret'])
auth.set_access_token(config['access_token'],
              config['access_token_secret'])
```

```python
api = tweepy.API(auth)
```

```python
data_tweets = api.user_timeline("analyticbridge", count=50)
```

```python
data_tweets
```

```python
tweets = []
for tweet in data_tweets:
    tweets.append([tweet.text, tweet.retweet_count, tweet.favorite_count])
print len(tweets)
```

```python
tweets
```

```python
datatweets = client.dsbc.datatweets
```

```python
for k in tweets:
    datatweets_d = {'id': k}
    datatweets.insert(datatweets_d)
```
