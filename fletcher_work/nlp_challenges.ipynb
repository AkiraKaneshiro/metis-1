```python
import math

def tf(word, docs):
    return docs.words.count(word) / len(docs.words)

def n_containing(word, docslist):
    return sum(1 for docs in docslist if word in docs)

def idf(word, docslist):
    return math.log(len(docslist) / (1 + n_containing(docs, docslist)))

def tfidf(word, docs, docslist):
    return tf(word, docs) * idf(word, docslist)
```

```python
docslist = documents
for i, docs in enumerate(docslist):
    docs = str(docs)
    print("Top words in document {}".format(i + 1))
    scores = {word: tfidf(word, docs, docslist) for word in docs.words}
    sorted_words = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    for word, score in sorted_words[:3]:
        print("\tWord: {}, TF-IDF: {}".format(word, round(score, 5)))
```
